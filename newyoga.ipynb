```python
yogas= pd.read_csv("C:\\Users\\U6067583\\Downloads\\project\\yogafinal.csv",parse_dates=['queryTime'])
```

```python
#creating a datframe with required columns
df1 = yogas[['searchTerms', 'rank','queryTime','title','displayLink','link','totalResults','count','formattedTotalResults']]
```

```python
#finding the mostsearched and most counted sites as per the given keywords
mostsearchedsites = (df1
           .groupby(['displayLink'], as_index=False)
           .agg({'rank': ['count', 'mean']})
           .sort_values(('rank', 'count'), ascending=False)
           .assign(coverage=lambda df: df[('rank', 'count')].div(df1['searchTerms'].nunique())))
mostsearchedsites.columns = ['displayLink', 'count', 'avg_rank', 'coverage']
mostsearchedsites['displayLink'] = mostsearchedsites['displayLink'].str.replace('www.', '')
mostsearchedsites['avg_rank'] = mostsearchedsites['avg_rank'].round(1)
mostsearchedsites['coverage'] = mostsearchedsites['coverage'].mul(100).round(1).astype(str).add('%')
mostsearchedsites.head(20).reset_index(drop=True)
```

```python
#ranking order
top_domains = df1['displayLink'].value_counts()[:10].index.tolist()
top_df = df1[df1['displayLink'].isin(top_domains)]

rank_counts = top_df.groupby(['displayLink', 'rank']).agg({'rank': ['count']}).reset_index()
rank_counts.columns = ['displayLink', 'rank', 'count']
rank_counts.head(10)
```
